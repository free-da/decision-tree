```python
import pandas as pd

df = pd.read_csv("data.csv", sep=';')
df
```

```python
#source: https://www.programcreek.com/python/?CodeExample=calculate+entropy

import numpy as np
import math

def calculate_entropy(y):
    """ Calculate the entropy of label array y """
    log2 = lambda x: math.log(x) / math.log(2)
    unique_labels = np.unique(y)
    entropy = 0
    for label in unique_labels:
        count = len(y[y == label])
        p = count / len(y)
        entropy += -p * log2(p)
    return entropy 

entropy_buys_computer = calculate_entropy(df['Buys_computer'])
entropy_buys_computer
```

```python
##entropy of age and buys_computer

def calculate_two_way_entropy(df):
    result = 0
    for option in ['Yes', 'No']:
        if df[option] > 0:
            p = df[option]/df['sum']
            result += -p * (math.log(p) / math.log(2))
    
    return result
```

```python
frequency_table_age = pd.crosstab(index=df['Age'], columns=df['Buys_computer'])
frequency_table_age
```

```python
information_gain = pd.DataFrame()
for attribute in df.columns:
    frequency_table = pd.crosstab(index=df[attribute], columns=df['Buys_computer'])
    frequency_table['sum'] = frequency_table.sum(axis=1)
    frequency_table['entropy'] = frequency_table.apply(lambda x: calculate_two_way_entropy(x), axis=1)
    frequency_table['p_of_yes'] = frequency_table.apply(lambda x: x['Yes']/frequency_table['Yes'].sum(axis=0), axis=1)
    frequency_table['entropy_proportion'] = frequency_table.apply(lambda x: x['entropy'] * x['p_of_yes'], axis=1)
    entropy_attribute = frequency_table['entropy_proportion'].sum(axis=0)
    information_gain[attribute] = [entropy_buys_computer - entropy_attribute]
    
information_gain
```
