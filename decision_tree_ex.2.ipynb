```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
 
df = pd.read_csv("data.csv", sep=';')

from sklearn.preprocessing import LabelEncoder

lb_make = LabelEncoder()
for attribute in ['Age', 'Income', 'Student', 'Credit_rating', 'Buys_computer']: 
    df[attribute] = lb_make.fit_transform(df[attribute])

X = df.iloc[:,[1,2,3,4]]
y = df['Buys_computer']
 
tree_clf = DecisionTreeClassifier(max_depth=3)
tree_clf.fit(X, y)
```

```python
df.iloc[:,[1,2,3,4]].columns
```

```python
from sklearn.tree import export_graphviz
    
export_graphviz(
    tree_clf,
    out_file="/Users/friederikebuchner/Documents/Beuth/Data Science/DET/computer_tree.dot",
    feature_names=df.iloc[:,[1,2,3,4]].columns,
    class_names=['Buys Computer', 'Buys no computer'],
    rounded=True,
    filled=True
)
```
